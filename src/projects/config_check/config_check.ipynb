## Verify the packages were installed successfylly
- To ensure that there are no hidden installation issues, we can invoke a few methods from the installed libraries to ensure they work as expected.
- Ignore the foreign nature of the syntax being used. The gist of the code is that we create a dataframe (you can think of it as an excel spreadsheet) and make a table with 1000 randomly generated numbers (`rnorm()`) for each `(x, y)` pair and then we pass the dataframe to `ggplot()`.

````r
df <- tibble(x=rnorm(1000), y=rnorm(1000))
ggplot(df, aes(x, y)) + geom_point(alpha=0.3)
````
